```python

# Leer el archivo CSV
file_path = path_data / 'datasets_preprocesados/sin regiones/combined_indicators.csv'
data = pd.read_csv(file_path)



# Agrupar datos por diferentes combinaciones de columnas y calcular la proporci√≥n de valores no nulos
grouped_data_file = data.groupby(['file_name', 'country_code', 'anho']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()
grouped_data_country = data.groupby(['country_code', 'indicator', 'anho']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()
grouped_data_anho = data.groupby(['anho', 'file_name', 'indicator']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()

# Crear los datasets de HoloViews
holo_data_file = hv.Dataset(grouped_data_file, ['anho', 'country_code'], 'proportion_not_null')
holo_data_country = hv.Dataset(grouped_data_country, ['anho', 'country_code'], 'proportion_not_null')
holo_data_anho = hv.Dataset(grouped_data_anho, ['anho', 'file_name'], 'proportion_not_null')

# Crear gr√°ficos interactivos
def create_plot(holo_data, groupby, title):
    plot = holo_data.to(hv.Curve, 'anho', 'proportion_not_null', groupby=groupby)
    plot.opts(
        opts.Curve(
            tools=['hover'],
            title=title,
            xlabel='A√±o',
            ylabel='Proporci√≥n de Valores No Nulos',
            height=400,
            width=600,
        ),
        opts.NdOverlay(legend_position='right')
    )
    return plot

# Men√∫ interactivo
selector = pn.widgets.Select(name='Agrupaci√≥n', options=['file_name', 'country_code', 'indicator'])

@pn.depends(selector.param.value)
def update_plot(selected):
    if selected == 'file_name':
        return create_plot(holo_data_file, 'file_name', "Proporci√≥n de Valores No Nulos por A√±o y Archivo")
    elif selected == 'country_code':
        return create_plot(holo_data_country, 'country_code', "Proporci√≥n de Valores No Nulos por A√±o y Pa√≠s")
    elif selected == 'indicator':
        return create_plot(holo_data_anho, 'indicator', "Proporci√≥n de Valores No Nulos por A√±o e Indicador")

# Layout
layout = pn.Column(selector, update_plot)

# Mostrar el layout
layout.servable()
```
