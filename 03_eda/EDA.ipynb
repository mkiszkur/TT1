```python

# Leer el archivo CSV
file_path = path_data / 'datasets_preprocesados/sin regiones/combined_indicators.csv'
data = pd.read_csv(file_path)



# Agrupar datos por diferentes combinaciones de columnas y calcular la proporción de valores no nulos
grouped_data_file = data.groupby(['file_name', 'country_code', 'anho']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()
grouped_data_country = data.groupby(['country_code', 'indicator', 'anho']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()
grouped_data_anho = data.groupby(['anho', 'file_name', 'indicator']).agg(proportion_not_null=('is_not_null', 'mean')).reset_index()

# Crear los datasets de HoloViews
holo_data_file = hv.Dataset(grouped_data_file, ['anho', 'country_code'], 'proportion_not_null')
holo_data_country = hv.Dataset(grouped_data_country, ['anho', 'country_code'], 'proportion_not_null')
holo_data_anho = hv.Dataset(grouped_data_anho, ['anho', 'file_name'], 'proportion_not_null')

# Crear gráficos interactivos
def create_plot(holo_data, groupby, title):
    plot = holo_data.to(hv.Curve, 'anho', 'proportion_not_null', groupby=groupby)
    plot.opts(
        opts.Curve(
            tools=['hover'],
            title=title,
            xlabel='Año',
            ylabel='Proporción de Valores No Nulos',
            height=400,
            width=600,
        ),
        opts.NdOverlay(legend_position='right')
    )
    return plot

# Menú interactivo
selector = pn.widgets.Select(name='Agrupación', options=['file_name', 'country_code', 'indicator'])

@pn.depends(selector.param.value)
def update_plot(selected):
    if selected == 'file_name':
        return create_plot(holo_data_file, 'file_name', "Proporción de Valores No Nulos por Año y Archivo")
    elif selected == 'country_code':
        return create_plot(holo_data_country, 'country_code', "Proporción de Valores No Nulos por Año y País")
    elif selected == 'indicator':
        return create_plot(holo_data_anho, 'indicator', "Proporción de Valores No Nulos por Año e Indicador")

# Layout
layout = pn.Column(selector, update_plot)

# Mostrar el layout
layout.servable()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
import os


# Crear el directorio de salida si no existe
output_dir = '/Users/miguelkiszkurno/Documents/TT1/data/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Ruta del archivo PDF de salida
pdf_path = os.path.join(output_dir, 'line_plots.pdf')

# Obtener las listas únicas de países e indicadores
countries = df_agriculture['country_name'].unique()
countries = df_agriculture['country_name'].unique()[:5] 
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])[:5]

df = df_agriculture
# Rango de años en el eje x
years = range(1960, 2023)

# Crear un archivo PDF y agregar gráficos de líneas
with PdfPages(pdf_path) as pdf:
    for country in countries:
        for indicator in indicators:
            filtered_df = df[df['country_name'] == country][['Año', indicator]].set_index('Año').reindex(years).reset_index()

            plt.figure(figsize=(10, 6))
            plt.plot(filtered_df['Año'], filtered_df[indicator], marker='o')
            plt.title(f'{indicator} en {country}')
            plt.xlabel('Año')
            plt.ylabel(indicator)
            plt.xlim(1960, 2022)
            plt.grid(True)
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            # Guardar la figura en el PDF
            pdf.savefig()
            plt.close()
print(f'Gráficos guardados en {pdf_path}')
```
