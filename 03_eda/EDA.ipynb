```python
# Contar la cantidad de indicadores por título de sección
section_counts = df_indicadores.groupby('titulo_seccion')['codigo_indicador'].count()

# Crear el diagrama de barras
plt.figure(figsize=(14, 8))
section_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Cantidad de Indicadores por Sección')
plt.xlabel('Título de Sección')
plt.ylabel('Cantidad de Indicadores')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
repeated_indicators = df_indicadores[df_indicadores.duplicated('codigo_indicador', keep=False)]
repeated_grouped = repeated_indicators.groupby('codigo_indicador')['titulo_seccion'].apply(list).reset_index()

repeated_grouped.columns = ['codigo_indicador', 'titulo_seccion_list']
repeated_grouped


# Filtrar los indicadores que se repiten 2 o más veces
repeated_indicators_count = repeated_indicators.groupby('codigo_indicador').size().reset_index(name='counts')
repeated_indicators_count = repeated_indicators_count[repeated_indicators_count['counts'] >= 2]


# Seleccionar los 20 indicadores más repetidos
top_repeated_indicators = repeated_indicators_count.nlargest(20, 'counts')

# Crear el gráfico de barras para los 20 indicadores más repetidos
plt.figure(figsize=(14, 14))
top_repeated_indicators.plot(kind='bar', x='codigo_indicador', y='counts', color='orange', edgecolor='black', legend=False)

plt.title('Top 20 Indicadores Repetidos y su Frecuencia')
plt.xlabel('Código del Indicador')
plt.ylabel('Cantidad de Repeticiones')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

## Paises

- High income group aggregate. High-income economies are those in which 2022 GNI per capita was more than $13,845.
- Low income group aggregate. Low-income economies are those in which 2022 GNI per capita was $1,135 or less.
- Lower middle income group aggregate. Lower-middle-income economies are those in which 2022 GNI per capita was between $1,136 and $4,465.
- Low and middle-income group aggregate. Low and middle-income economies are those in which 2022 GNI per capita was less than $13,845.

```python
df_paises = pd.read_csv(file_metadata_paises)
df_paises = df_paises[['Country Code', 'TableName', 'Region', 'IncomeGroup', 'SpecialNotes']]
df_paises.columns = ['country_code', 'country_name', 'region', 'income_group', 'notes']
df_paises

```

```python
# Obtener los valores únicos de 'country_code' y 'country_name'
unique_country_code = df_paises['country_code'].unique()
unique_country_name = df_paises['country_name'].unique()
unique_region_name = df_paises['region'].unique()
unique_income_group = df_paises['income_group'].unique()


# Contar los valores nulos por columna
null_values = df_paises.isnull().sum()

# Mostrar los resultados
unique_country_code, unique_country_name, null_values, unique_region_name, unique_income_group
```

```python
# Filtrar los registros que tienen 'region' e 'income_group' en null
null_region_income_group = df_paises[df_paises['region'].isnull() & df_paises['income_group'].isnull()]

null_region_income_group
```

```python
print(null_region_income_group['country_name'].unique().tolist())
```

```python
# Ordenar las categorías de 'income_group' para que aparezcan en el orden deseado
income_order = ['Low income', 'Lower middle income', 'Upper middle income', 'High income']
df_paises['income_group'] = pd.Categorical(df_paises['income_group'], categories=income_order, ordered=True)

# Filtrar los datos que no tienen valores nulos en 'region' y 'income_group'
df_filtered = df_paises.dropna(subset=['region', 'income_group'])

# Contar las ocurrencias de cada combinación de region e income_group
region_income_counts = df_filtered.groupby(['region', 'income_group']).size().unstack().fillna(0)

# Crear el gráfico de barras apiladas con el orden especificado
plt.figure(figsize=(14, 8))
region_income_counts[income_order].plot(kind='bar', stacked=True, colormap='tab20', edgecolor='black')

plt.title('Distribución de Grupos de Ingresos por Región')
plt.xlabel('Región')
plt.ylabel('Número de Países')
plt.legend(title='Grupo de Ingresos', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

## Agriculture & Rural Development

```python

df_agricultural_rural_dev = pd.read_csv(path_indicadores + 'Agriculture & Rural Development.csv')
df_agricultural_rural_dev = df_agricultural_rural_dev.rename(columns={'Country Code': 'country_code'})
df_agricultural_rural_dev = df_agricultural_rural_dev.rename(columns={'Country Name': 'country_name'})

df_agricultural_rural_dev
```

```python
# Realizar el merge de los DataFrames
df_merged = pd.merge(df_agricultural_rural_dev, df_paises[['country_code', 'region', 'income_group']], on='country_code', how='left')

df_agricultural_rural_dev = df_merged
# Mostrar las primeras filas del DataFrame resultante
df_agricultural_rural_dev[['country_code', 'country_name', 'region', 'income_group']]
```

```python
exclude_list = [
    'Africa Eastern and Southern', 'Africa Western and Central', 'Arab World',
    'Central Europe and the Baltics', 'Caribbean small states', 'East Asia & Pacific (excluding high income)',
    'Early-demographic dividend', 'East Asia & Pacific', 'Europe & Central Asia (excluding high income)',
    'Europe & Central Asia', 'Euro area', 'European Union', 'Fragile and conflict affected situations',
    'High income', 'Heavily indebted poor countries (HIPC)', 'IBRD only', 'IDA & IBRD total', 'IDA total',
    'IDA blend', 'IDA only', 'Latin America & Caribbean (excluding high income)', 'Latin America & Caribbean',
    'Least developed countries: UN classification', 'Low income', 'Lower middle income', 'Low & middle income',
    'Late-demographic dividend', 'Middle East & North Africa', 'Middle income',
    'Middle East & North Africa (excluding high income)', 'North America', 'OECD members', 'Other small states',
    'Pre-demographic dividend', 'Pacific island small states', 'Post-demographic dividend', 'South Asia',
    'Sub-Saharan Africa (excluding high income)', 'Sub-Saharan Africa', 'Small states', 'East Asia & Pacific (IDA & IBRD)',
    'Europe & Central Asia (IDA & IBRD)', 'Latin America & Caribbean (IDA & IBRD)', 'Middle East & North Africa (IDA & IBRD)',
    'South Asia (IDA & IBRD)', 'Sub-Saharan Africa (IDA & IBRD)', 'Upper middle income', 'World'
]

# Filtrar el DataFrame para excluir los registros de la lista
df_agricultural_rural_dev = df_agricultural_rural_dev[~df_agricultural_rural_dev['country_name'].isin(exclude_list)]

# Mostrar las primeras filas del DataFrame filtrado
df_agricultural_rural_dev
```

```python

# Cargar el archivo CSV
df_agriculture = df_agricultural_rural_dev

# Seleccionar las columnas que no son indicadores
non_indicator_columns = ['country_name', 'country_code', 'Año']

# Melt el DataFrame para tener una columna de indicadores y una columna de valores
df_melted = df_agriculture.melt(id_vars=non_indicator_columns, var_name='Indicator', value_name='Value')

# Contar los valores nulos por país e indicador
df_null_counts = df_melted.groupby(['country_name', 'Indicator'])['Value'].apply(lambda x: x.isnull().sum()).reset_index()

# Pivotear el DataFrame para obtener la matriz deseada
df_null_matrix = df_null_counts.pivot(index='country_name', columns='Indicator', values='Value').fillna(0)

# Mostrar la matriz resultante
print(df_null_matrix.head())
```

```python
# Transformar la matriz a formato largo
df_long = df_null_matrix.reset_index().melt(id_vars=['country_name'], var_name='Indicator', value_name='Cantidad_Nulos')
df_long

```

```python
# Filtrar por país e indicador
country = 'Argentina'  # Especificar el país deseado
indicator = 'AG.LND.AGRI.ZS'  # Especificar el indicador deseado

filtered_df = df_agriculture[df_agriculture['country_name'] == country]

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Año'], filtered_df[indicator], marker='o')

# Configurar el gráfico
plt.title(f'{indicator} en {country}')
plt.xlabel('Año')
plt.ylabel(indicator)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
import os


# Crear el directorio de salida si no existe
output_dir = '/Users/miguelkiszkurno/Documents/TT1/data/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Ruta del archivo PDF de salida
pdf_path = os.path.join(output_dir, 'line_plots.pdf')

# Obtener las listas únicas de países e indicadores
countries = df_agriculture['country_name'].unique()
countries = df_agriculture['country_name'].unique()[:5] 
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])[:5]

df = df_agriculture
# Rango de años en el eje x
years = range(1960, 2023)

# Crear un archivo PDF y agregar gráficos de líneas
with PdfPages(pdf_path) as pdf:
    for country in countries:
        for indicator in indicators:
            filtered_df = df[df['country_name'] == country][['Año', indicator]].set_index('Año').reindex(years).reset_index()

            plt.figure(figsize=(10, 6))
            plt.plot(filtered_df['Año'], filtered_df[indicator], marker='o')
            plt.title(f'{indicator} en {country}')
            plt.xlabel('Año')
            plt.ylabel(indicator)
            plt.xlim(1960, 2022)
            plt.grid(True)
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            # Guardar la figura en el PDF
            pdf.savefig()
            plt.close()
print(f'Gráficos guardados en {pdf_path}')
```
