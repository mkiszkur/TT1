```python
from matplotlib.backends.backend_pdf import PdfPages
import os


# Crear el directorio de salida si no existe
output_dir = '/Users/miguelkiszkurno/Documents/TT1/data/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Ruta del archivo PDF de salida
pdf_path = os.path.join(output_dir, 'line_plots.pdf')

# Obtener las listas únicas de países e indicadores
countries = df_agriculture['country_name'].unique()
countries = df_agriculture['country_name'].unique()[:5] 
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])
indicators = df_agriculture.columns.difference(['country_name', 'country_code', 'Año'])[:5]

df = df_agriculture
# Rango de años en el eje x
years = range(1960, 2023)

# Crear un archivo PDF y agregar gráficos de líneas
with PdfPages(pdf_path) as pdf:
    for country in countries:
        for indicator in indicators:
            filtered_df = df[df['country_name'] == country][['Año', indicator]].set_index('Año').reindex(years).reset_index()

            plt.figure(figsize=(10, 6))
            plt.plot(filtered_df['Año'], filtered_df[indicator], marker='o')
            plt.title(f'{indicator} en {country}')
            plt.xlabel('Año')
            plt.ylabel(indicator)
            plt.xlim(1960, 2022)
            plt.grid(True)
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            # Guardar la figura en el PDF
            pdf.savefig()
            plt.close()
print(f'Gráficos guardados en {pdf_path}')
```
