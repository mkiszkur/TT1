```python

def filtrar_dataset_por_anho_y_alpha(data, anho, alpha):
    df_indicators = data.copy()
    df_indicators = df_indicators[df_indicators['anho'] <= 2022]
    df_indicators = df_indicators[df_indicators['anho'] >= anho]

    # Convertir la columna 'anho' a tipo entero
    df_indicators['anho'] = df_indicators['anho'].astype(int)
    indicators = pd.DataFrame(df_indicators.groupby('indicator').filter(lambda x: (x['ratio_ones'] >= alpha).all())['indicator'].unique())
    indicators.columns = ['code']

    indicadores_path = path_data / 'merged_resultados_indicadores.csv'
    
    df_indicadores = pd.read_csv(indicadores_path)

    # Realizar el merge usando el campo codigo_indicador en resultados y INDICATOR_CODE en indicadores
    merged_data = pd.merge(indicators, df_indicadores, left_on='code', right_on='INDICATOR_CODE', how='inner')
    merged_data.to_csv(path_data / f'indicadores_filtrados_{anho}.csv', index=False)
    return merged_data

d = filtrar_dataset_por_anho_y_alpha(data, 2022, 0.95)
d
```

```python
import pandas as pd

pop_code = 'SP.POP.TOTL'
gni_code = 'NY.GNP.MKTP.CD'
anho_cluster = 2022

# Cargar los datasets
gni_path = path_data / 'csv' /  f'{gni_code}.csv'
pop_path = path_data / 'csv' /  f'{pop_code}.csv'
metadata_path =  path_data / 'metadata_countries.csv'

gni_data = pd.read_csv(gni_path)
pop_data = pd.read_csv(pop_path)
metadata = pd.read_csv(metadata_path)


# Renombrar las columnas para que sean consistentes
gni_data.columns = ['Country Name', 'Country Code', 'Year', 'GNI']
pop_data.columns = ['Country Name', 'Country Code', 'Year', 'Population']
metadata.columns = ['Country Code', 'Region', 'Income Group', 'special_notes', 'Country Name']

# Unificar los datasets por 'Country Name', 'Country Code', y 'Year'
merged_data = pd.merge(gni_data, pop_data, on=['Country Name', 'Country Code', 'Year'])
merged_data = pd.merge(merged_data, metadata, on=['Country Name', 'Country Code'])
merged_data.head()
```

V-Measure: Es la media armÃ³nica de la homogeneidad y la completitud. Los valores oscilan entre 0 y 1, donde 1 indica una coincidencia perfecta.
