### Obtener los datos

```python
import pandas as pd

pop_code = 'SP.POP.TOTL'
gni_code = 'NY.GNP.MKTP.CD'
anho_cluster = 2022

def get_data():
    # Cargar los datasets
    gni_path = path_data / 'csv' /  f'{gni_code}.csv'
    pop_path = path_data / 'csv' /  f'{pop_code}.csv'
    metadata_path =  path_data / 'metadata_countries.csv'

    gni_data = pd.read_csv(gni_path)
    pop_data = pd.read_csv(pop_path)
    metadata = pd.read_csv(metadata_path)


    # Renombrar las columnas para que sean consistentes
    gni_data.columns = ['Country Name', 'Country Code', 'Year', 'GNI']
    pop_data.columns = ['Country Name', 'Country Code', 'Year', 'Population']
    metadata.columns = ['Country Code', 'Region', 'Income Group', 'special_notes', 'Country Name']

    # Unificar los datasets por 'Country Name', 'Country Code', y 'Year'
    merged_data = pd.merge(gni_data, pop_data, on=['Country Name', 'Country Code', 'Year'])
    merged_data = pd.merge(merged_data, metadata, on=['Country Name', 'Country Code'])
    merged_data.head()
    return merged_data

def calcular_campos(df):
    df['gni_index'] = df['GNI'] / df['Population']
    df['log_gni_index'] = np.log1p(df['gni_index'])  # np.log1p para log(1 + x), que maneja mejor valores pequeÃ±os

    return df
def filtrar_nulos(df, campo):
    df = df.dropna(subset=[campo])
    return df

def filtrar_anho(df, anho):
    
    # Filtrar los datos para el aÃ±o 2022
    data = df[df['Year'] == anho]
    return data

def filtrar_regiones(df):
    df = df[~df['Country Code'].isin(excluir)]
    return df
```

```python
data_2022 = get_data()
data_2022 = calcular_campos(data_2022)
data_2022 = filtrar_anho(data_2022, anho_cluster)
data_2022 = filtrar_regiones(data_2022)
```

```python



def plot_gni_distribution(data, column, bins=50, figsize=(12, 8), thresholds=None, title='Distribution of GNI per Capita for 2022'):
    """
    FunciÃ³n para visualizar la distribuciÃ³n del GNI per Capita.

    :param data: DataFrame que contiene los datos.
    :param column: Columna del DataFrame que contiene los datos de GNI per Capita.
    :param bins: NÃºmero de bins para el histograma.
    :param figsize: TamaÃ±o de la figura.
    :param thresholds: Lista de umbrales para las lÃ­neas divisorias en formato [(valor, color, etiqueta)].
    :param title: TÃ­tulo del grÃ¡fico.
    """
    plt.figure(figsize=figsize)
    plt.hist(data[column].dropna(), bins=bins, edgecolor='k', alpha=0.7)
    plt.xlabel('GNI per Capita')
    plt.ylabel('Number of Countries')
    plt.title(title)
    
    if thresholds:
        for threshold, color, label in thresholds:
            plt.axvline(x=threshold, color=color, linestyle='--', label=label)
    
    plt.legend()
    plt.show()

# Definir los umbrales para las lÃ­neas divisorias
thresholds = [
    (1135, 'r', 'Low income / Lower middle income'),
    (4465, 'y', 'Lower middle income / Upper middle income'),
    (13845, 'g', 'Upper middle income / High income')
]

# Llamar a la funciÃ³n para visualizar la distribuciÃ³n del GNI per Capita
plot_gni_distribution(data_2022, 'gni_index', bins=50, thresholds=thresholds)
```

```python

def plot_gni_distributions(data, columns, bins=50, figsize=(18, 6), thresholds=None, titles=None):
    """
    FunciÃ³n para visualizar la distribuciÃ³n del GNI per Capita antes y despuÃ©s de la transformaciÃ³n logarÃ­tmica.

    :param data: DataFrame que contiene los datos.
    :param columns: Lista de columnas del DataFrame que contienen los datos de GNI per Capita y su transformaciÃ³n logarÃ­tmica.
    :param bins: NÃºmero de bins para el histograma.
    :param figsize: TamaÃ±o de la figura.
    :param thresholds: Lista de umbrales para las lÃ­neas divisorias en formato [(valor, color, etiqueta)].
    :param titles: Lista de tÃ­tulos para cada grÃ¡fico.
    """
    fig, axes = plt.subplots(1, len(columns), figsize=figsize)

    for i, ax in enumerate(axes):
        ax.hist(data[columns[i]].dropna(), bins=bins, edgecolor='k', alpha=0.7)
        ax.set_xlabel('GNI per Capita' if 'log' not in columns[i] else 'Log of GNI per Capita')
        ax.set_ylabel('Number of Countries')
        ax.set_title(titles[i])
        
        if thresholds:
            for threshold, color, label in thresholds:
                ax.axvline(x=threshold if 'log' not in columns[i] else np.log1p(threshold), color=color, linestyle='--', label=label)
        
        ax.legend()

    plt.tight_layout()
    plt.show()

# Definir los umbrales para las lÃ­neas divisorias
thresholds = [
    (1135, 'r', 'Low income / Lower middle income'),
    (4465, 'y', 'Lower middle income / Upper middle income'),
    (13845, 'g', 'Upper middle income / High income')
]

# Definir las columnas y tÃ­tulos para los histogramas
columns = ['gni_index', 'log_gni_index']
titles = ['Distribution of GNI per Capita for 2022', 'Distribution of Log of GNI per Capita for 2022']

# Llamar a la funciÃ³n para visualizar las distribuciones
plot_gni_distributions(data_2022, columns, bins=50, thresholds=thresholds, titles=titles)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def visualize_clusters(data, cluster_col, gni_col, country_col, cluster_mapping, colors, thresholds, path_to_save=None):
    """
    FunciÃ³n para visualizar los clusters de paÃ­ses por GNI per Capita.

    :param data: DataFrame que contiene los datos.
    :param cluster_col: Columna que contiene los clusters.
    :param gni_col: Columna que contiene los datos de GNI per Capita.
    :param country_col: Columna que contiene los nombres de los paÃ­ses.
    :param cluster_mapping: Diccionario que mapea los nÃºmeros de cluster a las categorÃ­as de ingreso.
    :param colors: Diccionario que mapea los nÃºmeros de cluster a los colores.
    :param thresholds: Lista de umbrales para las lÃ­neas divisorias en formato [(valor, color, etiqueta)].
    :param path_to_save: Ruta para guardar el DataFrame con los clusters mapeados (opcional).
    """
    # Mapear los nÃºmeros de cluster a las categorÃ­as de ingreso
    data['Cluster Mapped'] = data[cluster_col].map(cluster_mapping)
    
    # Guardar el DataFrame con los clusters mapeados si se proporciona una ruta
    if path_to_save:
        data.to_csv(path_to_save, index=False)

    # Asignar colores a cada punto del scatter plot
    cluster_colors = data[cluster_col].map(colors)

    # VisualizaciÃ³n de los clusters
    plt.figure(figsize=(12, 8))
    plt.scatter(data[country_col], data[gni_col], c=cluster_colors)
    plt.xticks(rotation=90)
    plt.xlabel('Country')
    plt.ylabel('GNI per Capita')
    plt.title('Clustering of Countries by GNI per Capita for 2022')
    
    # AÃ±adir las lÃ­neas divisorias
    for threshold, color, label in thresholds:
        plt.axhline(y=threshold, color=color, linestyle='--', label=label)
    
    # Agregar una leyenda personalizada usando cluster_mapping
    legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[cluster], markersize=10, label=label)
                      for cluster, label in cluster_mapping.items()]
    
    plt.legend(handles=legend_handles, title='Cluster')
    plt.legend()
    plt.show()


# Definir el mapeo de clusters y colores
cluster_mapping = {2: 'Upper middle income', 1: 'High income', 3: 'Low income', 0: 'Lower middle income'}
colors = {1: 'green', 2: 'yellow', 0: 'orange', 3: 'red'}

# Definir los umbrales para las lÃ­neas divisorias
thresholds = [
    (1135, 'r', 'Low income / Lower middle income'),
    (4465, 'y', 'Lower middle income / Upper middle income'),
    (13845, 'g', 'Upper middle income / High income')
]

# Llamar a la funciÃ³n para visualizar los clusters
visualize_clusters(data_2022, 'Cluster', 'gni_index', 'Country Name', cluster_mapping, colors, thresholds, path_to_save='clusters.csv')
```

V-Measure: Es la media armÃ³nica de la homogeneidad y la completitud. Los valores oscilan entre 0 y 1, donde 1 indica una coincidencia perfecta.
