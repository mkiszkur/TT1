```python
# Definir la función para crear el layout
def create_heatmap_dashboard(file_path, filter_by_file_name=True, filtro_indicadores = None, show_labels = True):
    # Cargar el dataset
    
    data = pd.read_csv(file_path)
    
    # Filtra por la lista de indicadores que quiero ver
    if filtro_indicadores != None:
        data = data[data['indicator'].isin(filtro_indicadores)]
        

    # Crear un selector para file_name si es necesario
    if filter_by_file_name:
        file_names = list(data['file_name'].unique())
        file_name_select = pn.widgets.Select(name='File Name', options=file_names)
    else:
        file_name_select = None

    # Crear un selector para indicators
    indicators = list(data['indicator'].unique())
    indicator_select = pn.widgets.MultiSelect(name='Indicators', options=indicators, size=10)

    # Crear un slider para alpha
    alpha_slider = pn.widgets.FloatSlider(name='Alpha', start=0, end=1, step=0.01, value=0.5)

    # Crear un rango para los años
    min_year = data['anho'].min()
    max_year = data['anho'].max()
    year_slider = pn.widgets.IntRangeSlider(name='Años', start=min_year, end=max_year, step=1, value=(min_year, max_year))

    # Función para actualizar el heatmap
    if filter_by_file_name:
        @pn.depends(file_name_select.param.value, indicator_select.param.value, alpha_slider.param.value, year_slider.param.value)
        def update_heatmap(file_name, selected_indicators, alpha, year_range):
            filtered_data = data[data['file_name'] == file_name]
            if selected_indicators:
                filtered_data = filtered_data[filtered_data['indicator'].isin(selected_indicators)]
            
            filtered_data = filtered_data[filtered_data['ratio_ones'] >= alpha]
            filtered_data = filtered_data[(filtered_data['anho'] >= year_range[0]) & (filtered_data['anho'] <= year_range[1])]
            ylabel = 'Indicador' if show_labels else ''            
            heatmap = hv.HeatMap(filtered_data, kdims=['anho', 'indicator'], vdims=['ratio_ones'])
            heatmap = heatmap.opts(
                cmap='Viridis', 
                colorbar=True, 
                width=900, 
                height=600, 
                xlabel='Año', 
                ylabel=ylabel, 
                tools=['hover'],
                xrotation=45
            )
            return heatmap

        # Layout
        layout = pn.Column(
            pn.Row(file_name_select, indicator_select, alpha_slider, year_slider),
            update_heatmap
        )
    else:
        @pn.depends(indicator_select.param.value, alpha_slider.param.value, year_slider.param.value)
        def update_heatmap(selected_indicators, alpha, year_range):
            filtered_data = data
            if selected_indicators:
                filtered_data = filtered_data[filtered_data['indicator'].isin(selected_indicators)]
            filtered_data = filtered_data[filtered_data['ratio_ones'] >= alpha]
            filtered_data = filtered_data[(filtered_data['anho'] >= year_range[0]) & (filtered_data['anho'] <= year_range[1])]
            ylabel = 'Indicador' if show_labels else ''                   
            heatmap = hv.HeatMap(filtered_data, kdims=['anho', 'indicator'], vdims=['ratio_ones'])
            heatmap = heatmap.opts(
                cmap='Viridis', 
                colorbar=True, 
                width=900, 
                height=600, 
                xlabel='Año', 
                ylabel=ylabel, 
                tools=['hover'],
                xrotation=45
            )
            return heatmap

        # Layout
        layout = pn.Column(
            pn.Row(indicator_select, alpha_slider, year_slider),
            update_heatmap
        )

    return layout
```

```python


def cantidad_indicadores (alpha, anho_desde, anho_hasta, df):

    df = df[df['anho'] <= anho_hasta]

    # Filtrar los datos y contar los indicadores válidos año por año
    for x in range(anho_desde, anho_hasta):
        # Filtrar los datos para incluir solo los años desde x en adelante
        df = df[df['anho'] >= x]

        # Agrupar por indicador y filtrar aquellos que cumplen la condición en todos sus registros
        indicators = df.groupby('indicator').filter(lambda x: (x['ratio_ones'] >= alpha).all())['indicator'].unique()
        
        
        # Contar el número de indicadores únicos que cumplen con la condición
        cant = indicators.shape[0]
        
        # Imprimir el resultado
        print(f'{x}: {cant}')

```

```python
def filtrar_dataset_por_anho_y_alpha(data, anho, alpha):
    df_indicators = data.copy()
    df_indicators = df_indicators[df_indicators['anho'] <= 2022]
    df_indicators = df_indicators[df_indicators['anho'] >= anho]

    # Convertir la columna 'anho' a tipo entero
    df_indicators['anho'] = df_indicators['anho'].astype(int)
    indicators = pd.DataFrame(df_indicators.groupby('indicator').filter(lambda x: (x['ratio_ones'] >= alpha).all())['indicator'].unique())
    indicators.columns = ['code']

    indicadores_path = path_data / 'merged_resultados_indicadores.csv'
    
    df_indicadores = pd.read_csv(indicadores_path)

    # Realizar el merge usando el campo codigo_indicador en resultados y INDICATOR_CODE en indicadores
    merged_data = pd.merge(indicators, df_indicadores, left_on='code', right_on='INDICATOR_CODE', how='inner')
    merged_data[['titulo_seccion', 'codigo_indicador', 'INDICATOR_NAME', 'estado']].to_csv(path_data / f'indicadores_filtrados_{anho}.csv', index=False)
    return merged_data
```

```python
def plot_indicators_seaborn(data, indicators=None, countries=None):
    # Filtrar los datos según los parámetros indicados
    if indicators is not None:
        data = data[['country_name', 'anho'] + indicators]
    if countries is not None:
        data = data[data['country_name'].isin(countries)]
    
    # Crear un DataFrame largo para seaborn
    data_melted = data.melt(id_vars=['country_name', 'anho'], value_vars=indicators, 
                            var_name='Indicador', value_name='Valor')
    
    # Crear widgets para seleccionar el país y el indicador
    indicator_options = indicators if indicators else data.columns[3:].tolist()
    country_options = countries if countries else data['country_name'].unique().tolist()
    
    indicator_select = widgets.Dropdown(options=indicator_options, description='Indicador:')
    country_select = widgets.Dropdown(options=country_options, description='País:')
    
    def update_plot(change):
        selected_indicator = indicator_select.value
        selected_country = country_select.value
        
        filtered_data = data_melted[(data_melted['Indicador'] == selected_indicator) & 
                                    (data_melted['country_name'] == selected_country)]
        
        # Limpiar la salida anterior
        clear_output(wait=True)
        
        # Volver a mostrar los widgets
        display(indicator_select, country_select)
        
        # Crear la figura y el gráfico
        plt.figure(figsize=(14, 8))
        sns.lineplot(data=filtered_data, x='anho', y='Valor', hue='Indicador', style='country_name', markers=True, dashes=False)
        
        plt.title(f'Evolución de {selected_indicator} en {selected_country}')
        plt.xlabel('Año')
        plt.ylabel('Valor')
        plt.legend(title='Indicador / País', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        
        plt.show()
    
    indicator_select.observe(update_plot, names='value')
    country_select.observe(update_plot, names='value')
    
    display(indicator_select, country_select)
    
    # Graficar inicialmente con los valores por defecto
    update_plot(None)
```

```python
def aplicar_clustering(df, cant_clusters, campo):
    # Escalar los datos de gni_index
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(df[[campo]])

    # Aplicar k-means clustering
    kmeans = KMeans(n_clusters=cant_clusters, random_state=42)
    df['Cluster'] = kmeans.fit_predict(data_scaled)
    
    # Calcular el coeficiente de silhouette
    silhouette_avg = silhouette_score(data_scaled, df['Cluster'])
    
    # Calcular la inercia (suma de las distancias al cuadrado entre cada punto y el centroide de su cluster)
    inertia = kmeans.inertia_
    
    return df, silhouette_avg, inertia
```

```python


def visualize_clusters(data, cluster_col, gni_col, country_col, country_code_col, target_country_codes, cluster_mapping, colors, thresholds, show_xticks = False, path_to_save=None):
    """
    Función para visualizar los clusters de países por GNI per Capita.
    
    :param data: DataFrame que contiene los datos.
    :param cluster_col: Columna que contiene los clusters.
    :param gni_col: Columna que contiene los datos de GNI per Capita.
    :param country_col: Columna que contiene los nombres de los países.
    :param country_code_col: Columna que contiene los códigos de los países.
    :param target_country_codes: Lista de códigos de países que se desea etiquetar.
    :param cluster_mapping: Diccionario que mapea los números de cluster a las categorías de ingreso.
    :param colors: Diccionario que mapea los números de cluster a los colores.
    :param thresholds: Lista de umbrales para las líneas divisorias en formato [(valor, color, etiqueta)].
    :param path_to_save: Ruta para guardar el DataFrame con los clusters mapeados (opcional).
    """
    # Eliminar filas con NaN en las columnas relevantes
    data = data.dropna(subset=[cluster_col, gni_col, country_col, country_code_col])
    
    # Convertir las columnas a los tipos adecuados
    data[country_col] = data[country_col].astype(str)
    data[gni_col] = pd.to_numeric(data[gni_col], errors='coerce')
    data = data.dropna(subset=[gni_col])  # Eliminar filas donde gni_col no se pudo convertir

    # Mapear los números de cluster a las categorías de ingreso
    data['Cluster Mapped'] = data[cluster_col].map(cluster_mapping)
    
    # Guardar el DataFrame con los clusters mapeados si se proporciona una ruta
    if path_to_save:
        data.to_csv(path_to_save, index=False)

    # Asignar colores a cada punto del scatter plot
    cluster_colors = data[cluster_col].map(colors)

    # Visualización de los clusters
    plt.figure(figsize=(12, 8))
    plt.scatter(data[country_col], data[gni_col], c=cluster_colors)
    if show_xticks:
        plt.xticks(rotation=90)
    else:
        plt.xticks([])
        plt.gca().tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    #plt.xticks(rotation=90)
    plt.xlabel('Country')
    plt.ylabel('GNI per Capita')
    plt.title('Clustering of Countries by GNI per Capita for 2022')
    
    # Etiquetar los países con los códigos proporcionados
    target_countries = data[data[country_code_col].isin(target_country_codes)]
    for _, row in target_countries.iterrows():
        plt.annotate(row[country_col], (row[country_col], row[gni_col]), textcoords="offset points", xytext=(-30,10), ha='center', fontsize=10, color='black', fontweight='bold',
                     arrowprops=dict(arrowstyle="->", color='black', lw=1))
    
    # Añadir las líneas divisorias
    for threshold, color, label in thresholds:
        plt.axhline(y=threshold, color=color, linestyle='--', label=label)
    
    # Agregar una leyenda personalizada usando cluster_mapping
    legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[cluster], markersize=10, label=label)
                      for cluster, label in cluster_mapping.items()]
    
    plt.legend(handles=legend_handles, title='Cluster')
    plt.legend()
    plt.show()
```

```python
def comparar_medias(df, agrupador, indicador):
    # Cargar el nuevo dataset
    data_income_groups = df



    # Calcular las medias del gni_index para cada income group
    income_group_means = data_income_groups.groupby(agrupador)[indicador].mean()

    # Calcular la desviación estándar global del gni_index
    std_gni_index = data_income_groups['gni_index'].std()

    # Calcular las distancias entre las medias de los income groups en términos de desviaciones estándar
    distances = []

    for i in range(len(income_group_means)):
        for j in range(i + 1, len(income_group_means)):
            group1 = income_group_means.index[i]
            group2 = income_group_means.index[j]
            distance = abs(income_group_means[i] - income_group_means[j]) / std_gni_index
            distances.append([group1, group2, distance])

    income_group_distances = pd.DataFrame(distances, columns=['Group 1', 'Group 2', 'Distance (std devs)'])


    return income_group_means, income_group_distances
```

```python
# Definir la función para crear el layout
def create_heatmap_dashboard_area_country_anho(file_path, filter_by_file_name=True, filtro_indicadores = None, show_labels = True):
    # Cargar el dataset
    
    data = pd.read_csv(file_path)
    
    # Filtra por la lista de indicadores que quiero ver
    if filtro_indicadores != None:
        data = data[data['country_code'].isin(filtro_indicadores)]
        

    # Crear un selector para file_name si es necesario
    if filter_by_file_name:
        file_names = list(data['file_name'].unique())
        file_name_select = pn.widgets.Select(name='File Name', options=file_names)
    else:
        file_name_select = None

    # Crear un selector para paises
    paises = list(data['country_code'].unique())
    country_code_select = pn.widgets.MultiSelect(name='paises', options=paises, size=10)

    # Crear un slider para alpha
    alpha_slider = pn.widgets.FloatSlider(name='Alpha', start=0, end=1, step=0.01, value=0.5)

    # Crear un rango para los años
    min_year = data['anho'].min()
    max_year = data['anho'].max()
    year_slider = pn.widgets.IntRangeSlider(name='Años', start=min_year, end=max_year, step=1, value=(min_year, max_year))

    
    # Función para actualizar el heatmap
    if filter_by_file_name:
        @pn.depends(file_name_select.param.value, country_code_select.param.value, alpha_slider.param.value, year_slider.param.value)
        def update_heatmap(file_name, selected_paises, alpha, year_range):
            filtered_data = data[data['file_name'] == file_name]
            if selected_paises:
                filtered_data = filtered_data[filtered_data['country_code'].isin(selected_paises)]
            
            filtered_data = filtered_data[filtered_data['ratio_ones'] >= alpha]
            filtered_data = filtered_data[(filtered_data['anho'] >= year_range[0]) & (filtered_data['anho'] <= year_range[1])]
            ylabel = 'Indicador' if show_labels else ''            
            heatmap = hv.HeatMap(filtered_data, kdims=['anho', 'country_code'], vdims=['ratio_ones'])
            heatmap = heatmap.opts(
                cmap='Viridis', 
                colorbar=True, 
                width=900, 
                height=600, 
                xlabel='Año', 
                ylabel=ylabel, 
                tools=['hover'],
                xrotation=45
            )
            return heatmap

        # Layout
        layout = pn.Column(
            pn.Row(file_name_select, country_code_select, alpha_slider, year_slider),
            update_heatmap
        )
    else:
        @pn.depends(country_code_select.param.value, alpha_slider.param.value, year_slider.param.value)
        def update_heatmap(selected_paises, alpha, year_range):
            filtered_data = data
            if selected_paises:
                filtered_data = filtered_data[filtered_data['country_code'].isin(selected_paises)]
            filtered_data = filtered_data[filtered_data['ratio_ones'] >= alpha]
            filtered_data = filtered_data[(filtered_data['anho'] >= year_range[0]) & (filtered_data['anho'] <= year_range[1])]
            ylabel = 'Indicador' if show_labels else ''                   
            heatmap = hv.HeatMap(filtered_data, kdims=['anho', 'country_code'], vdims=['ratio_ones'])
            heatmap = heatmap.opts(
                cmap='Viridis', 
                colorbar=True, 
                width=900, 
                height=600, 
                xlabel='Año', 
                ylabel=ylabel, 
                tools=['hover'],
                xrotation=45
            )
            return heatmap

        # Layout
        layout = pn.Column(
            pn.Row(country_code_select, alpha_slider, year_slider),
            update_heatmap
        )

    return layout
```

```python
def merge_files_with_metadata(file_list, metadata_file):
    """
    Combina múltiples archivos CSV en uno y luego lo une con un archivo de metadatos.
    
    Parameters:
    - file_list: Lista de rutas a archivos CSV que deben ser combinados.
    - metadata_file: Ruta al archivo CSV de metadatos.
    
    Returns:
    - DataFrame resultante de la combinación de todos los archivos y los metadatos.
    """
    # Leer el archivo de metadatos
    metadata = pd.read_csv(metadata_file)
    
    # Inicializar el DataFrame combinado
    combined_df = None
    
    # Leer y combinar todos los archivos en file_list
    for file in file_list:
        df = pd.read_csv(file)
        
        if combined_df is None:
            combined_df = df
        else:
            combined_df = pd.merge(combined_df, df, on=['country_code', 'Año'], how='outer')
        
            # Renombrar columnas duplicadas
            combined_df = combined_df.rename(columns={'country_code_x': 'country_code', 'country_name_x': 'country_name'})

    # Después de todos los merges, eliminar columnas duplicadas si existen
    columns_to_drop = ['country_code_y', 'country_name_y']
    for col in columns_to_drop:
        if col in combined_df.columns:
            combined_df = combined_df.drop(columns=[col])
    
    # Realizar el merge con el archivo de metadatos
    final_df = pd.merge(combined_df, metadata, on='country_code', how='left')
    final_df = final_df.rename(columns={'country_code_x': 'country_code', 'country_name_x': 'country_name'})
    final_df = final_df.drop(columns=['country_name_y', 'special_notes'])

    return final_df
def get_list_indicators_file_names (indicators):
    
    result = []
    for i in indicators:
        result.append (path_data / f'csv/{i}.csv')
    return result

def generar_archivo_compuesto (lista_indicadores, file_name = 'Indicators.csv'):

    metadata_file = path_data / 'metadata_countries.csv'
    result = merge_files_with_metadata(get_list_indicators_file_names(lista_indicadores), metadata_file)
    result.to_csv(path_data / file_name, index=False)
    return result
    
```

```python

def perform_kmeans_clustering(data, n_clusters=4, indicators = None):
    """
    Realiza clustering usando KMeans en los indicadores de datos.
    
    Parameters:
    - data: DataFrame con los datos combinados.
    - n_clusters: Número de clústeres para KMeans.
    
    Returns:
    - DataFrame con los datos originales y una columna adicional para los clústeres.
    - El modelo KMeans ajustado.
    """
    # Filtrar las columnas de indicadores
    if indicators == None:
        indicator_columns = [col for col in data.columns if col not in ['country_code', 'Año', 'country_name', 'region','income_group', 'special_notes']]
    else: 
        indicator_columns = indicators

    print (indicator_columns)    
    # Filtrar filas con datos completos en los indicadores seleccionados
    data_filtered = data.dropna(subset=indicator_columns)
    
    # Obtener los registros eliminados
    deleted_records = data[~data.index.isin(data_filtered.index)]

    
    # Estandarizar los datos
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data_filtered[indicator_columns])
    

    # Aplicar KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    data_filtered['cluster'] = kmeans.fit_predict(data_scaled)
    
    # Aplicar PCA
    pca = PCA(n_components=2)
    components = pca.fit_transform(data_scaled)
    pca_df = pd.DataFrame(data=components, columns=['PCA1', 'PCA2'])
    data_filtered = data_filtered.reset_index(drop=True)
    data_filtered = pd.concat([data_filtered, pca_df], axis=1)
    
    return data_filtered, kmeans, pca, deleted_records
```

```python
def imprimir_metricas_pca (pca, clustered_data):
    # Varianza explicada por cada componente principal
    explained_variance = pca.explained_variance_ratio_
    print(f"Varianza explicada por PCA1: {explained_variance[0]*100:.2f}%")
    print(f"Varianza explicada por PCA2: {explained_variance[1]*100:.2f}%")

    # Analizar las características dentro de cada clúster
    numeric_columns = clustered_data.select_dtypes(include=[np.number]).columns
    cluster_means = clustered_data.groupby('cluster')[numeric_columns].mean()
    print("Media de características por clúster:")
    print(cluster_means)

    # Varianza explicada acumulada
    cumulative_explained_variance = explained_variance.cumsum()
    print(f"Varianza explicada acumulada por PCA1 y PCA2: {cumulative_explained_variance[1]*100:.2f}%")
    
    return explained_variance, cumulative_explained_variance, cluster_means

```

Adjusted Rand Index (ARI): Mide la similitud entre dos agrupamientos, ajustando por la posibilidad de agrupamientos aleatorios. Los valores oscilan entre -1 y 1, donde 1 indica una coincidencia perfecta, 0 indica una coincidencia no mejor que el azar y valores negativos indican coincidencias peores que el azar.

Normalized Mutual Information (NMI): Mide la cantidad de información compartida entre dos agrupamientos. Los valores oscilan entre 0 y 1, donde 1 indica una coincidencia perfecta y 0 indica que no hay información compartida.

Homogeneity: Indica que todos los puntos de cada cluster pertenecen a la misma clase verdadera. Los valores oscilan entre 0 y 1, donde 1 indica homogeneidad perfecta.

Completeness: Indica que todos los puntos de una clase verdadera están asignados al mismo cluster. Los valores oscilan entre 0 y 1, donde 1 indica completitud perfecta.

V-Measure: Es la media armónica de la homogeneidad y la completitud. Los valores oscilan entre 0 y 1, donde 1 indica una coincidencia perfecta.

```python


def plot_pca_bokeh(data):
    """
    Crea una visualización de PCA utilizando Bokeh para permitir la interacción.
    
    Parameters:
    - data: DataFrame con los resultados de PCA y los clústeres.
    """
    # Crear un mapa de colores para los clústeres
    cluster_colors = Category10[len(data['cluster'].unique())]
    color_map = {str(i): cluster_colors[i] for i in range(len(data['cluster'].unique()))}
    
    # Agregar una columna de color al DataFrame
    data['color'] = data['cluster'].astype(str).map(color_map)
    
    # Agregar columna de tag
    tags = ['ARG', 'BMU', 'CHE', 'NOR']
    data['tagged_country'] = data.apply(lambda row: row['country_name'] if row['country_code'] in tags else '', axis=1)
   
   
    # Crear una fuente de datos para Bokeh
    source = ColumnDataSource(data={
        'PCA1': data['PCA1'],
        'PCA2': data['PCA2'],
        'country_name': data['country_name'],
        'cluster': data['cluster'].astype(str),
        'SL.TLF.TOTL.FE.ZS': data['SL.TLF.TOTL.FE.ZS'],
        'SL.TLF.TOTL.IN': data['SL.TLF.TOTL.IN'],
        'SL.UEM.TOTL.FE.ZS': data['SL.UEM.TOTL.FE.ZS'],
        'income_group': data['income_group'],
        'region': data['region'],
#        'SL.UEM.TOTL.MA.ZS': data['SL.UEM.TOTL.MA.ZS'],
#        'SL.UEM.TOTL.ZS': data['SL.UEM.TOTL.ZS'],
#        'SP.POP.TOTL': data['SP.POP.TOTL'],
#        'NY.GNP.MKTP.CD': data['NY.GNP.MKTP.CD'],
        'color': data['color']
    })
    
    # Crear la figura de Bokeh
    p = figure(title='Clustering de Indicadores de Educación y Fuerza Laboral (PCA)',
               x_axis_label='PCA1', y_axis_label='PCA2', width=800, height=600,
               tools="pan,wheel_zoom,box_zoom,reset,save")
    
    # Añadir círculos para cada punto de datos
    p.circle('PCA1', 'PCA2', source=source, size=10, color='color', legend_field='cluster')

    # Añadir anotaciones para los países específicos
    tags = ['ARG', 'BMU', 'CHE', 'NOR', 'BDI']
    for tag in tags:
        tagged_data = data[data['country_code'] == tag]
        for index, row in tagged_data.iterrows():
            label = Label(x=row['PCA1'], y=row['PCA2'], text=row['country_name'],
                          x_offset=10, y_offset=10, text_font_size='10pt', text_color='black')
            arrow = Arrow(end=VeeHead(size=10), x_start=row['PCA1'], y_start=row['PCA2'], 
                          x_end=row['PCA1'] + 0.1, y_end=row['PCA2'] + 0.1, line_width=2)
            p.add_layout(label)
            p.add_layout(arrow)
    
    # Añadir la herramienta de hover
    hover = HoverTool()
    hover.tooltips = [
        ("Region", "@region"),
        ("Income Group", "@income_group"),
        ("País", "@country_name"),
        ("Cluster", "@cluster"),
        ("SL.TLF.TOTL.FE.ZS", "@{SL.TLF.TOTL.FE.ZS}"),
        ("SL.TLF.TOTL.IN", "@{SL.TLF.TOTL.IN}"),
        ("SL.UEM.TOTL.FE.ZS", "@{SL.UEM.TOTL.FE.ZS}")
#        ("SL.UEM.TOTL.MA.ZS", "@{SL.UEM.TOTL.MA.ZS}"),
#        ("SL.UEM.TOTL.ZS", "@{SL.UEM.TOTL.ZS}"),
#        ("SP.POP.TOTL", "@{SP.POP.TOTL}"),
#        ("NY.GNP.MKTP.CD", "@{NY.GNP.MKTP.CD}"),
    ]
    p.add_tools(hover)
    
    # Configurar leyenda
    p.legend.title = 'Cluster'
    p.legend.location = 'top_right'
    
    # Mostrar la figura
    show(p)

```
